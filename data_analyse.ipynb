```python
def encode(lat: float, lng: float, level: int):
    if level < 1:
        return null
    res = ""
    if lng >= 0:
        res += "E"
    else:
        res += "W"
        lng += 180
    lat += 90 # 180:the North Pole,  0:the South Pole
    unit = 180
    for i in range(1, level):
        unit /= 3
        x = int(lng / unit)
        y = int(lat / unit)
        res += chr(ord('0') + x + y * 3 + 1)
        lng -= x * unit
        lat -= y * unit
    return res
```

```python
res = []
for idx, row in kunitaka_df.iterrows():
    res.append([row["事故内容"], row["事故類型"], row["年齢（当事者A）"], row["年齢（当事者B）"], row["当事者種別（当事者A）"], row["当事者種別（当事者B）"], encode(row["緯度"], row["経度"], 15)])
res = pd.DataFrame(res)
res = res.rename(columns={0: "事故内容", 1: "事故類型", 2: "年齢（当事者A）", 3: "年齢（当事者B）", 4: "当事者種別（当事者A）", 5: "当事者種別（当事者B）", 6: "Geo3x3" })
res
```
